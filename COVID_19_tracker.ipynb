Worldometers has a credible sources of COVID19 data.


In this exersice, we will learn how to get data from Worldmeters website
 and practice how to plot the data.


```python
import re
import requests
import pandas as pd

html_source = requests.get("https://www.worldometers.info/coronavirus").text
html_source = re.sub(r'<.*?>', lambda g: g.group(0).upper(), html_source)

df=pd.read_html(html_source)
df
```

```python
df=df[0]
```

```python
df.head(30)
```

```python
df.info()
```

```python
df=df.drop('#',axis=1)
```

```python
df.head(10)
```

```python
#Removing NAN values so we can plot our data correctly
clean=df.dropna()
```

```python
#showing total cases as a warning
clean.sort_values('TotalCases',inplace=True,ascending=True)
clean.plot(kind='barh',x='Country,Other',y='TotalCases',color='orange',figsize=(15,15))
```

```python
#showing total deaths per country
clean.sort_values('TotalDeaths',inplace=True,ascending=True)
clean.plot(kind='barh',x='Country,Other',y='TotalDeaths',color='black',figsize=(15,15))
```

```python
#showing total recovered cases as hope
clean.sort_values('TotalRecovered',inplace=True,ascending=True)
clean.plot(kind='barh',x='Country,Other',y='TotalRecovered',color='g',figsize=(15,15))
```

```python
#showing critical cases as an alert
clean.sort_values('Serious,Critical',inplace=True,ascending=True)
clean.plot(kind='barh',x='Country,Other',y='Serious,Critical',color='r',figsize=(15,15))
```
